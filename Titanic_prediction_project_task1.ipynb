```python
#Data transformation,Preprocessing into dataframes
df = pd.read_csv('Titanic-Dataset.csv')
headers = df.columns
split_data = df.iloc[:,0].astype(str).str.split(',', expand=True)
df = pd.concat([split_data], axis=1).drop([0])

#print(df.columns)
df.columns = headers
df['FirstName']=df['FirstName'].str.replace('"','',regex=True)
df['LastName']=df['LastName'].replace('"','',regex=True)

df.head(50)

#Data Cleaning
#Checking for missing values

missing_values = df.isnull().sum()
print(missing_values)

#Save the data into a new csv file
df.to_csv('Titanic-Dataset-clean.csv', index=False)

df.head(50)



```
